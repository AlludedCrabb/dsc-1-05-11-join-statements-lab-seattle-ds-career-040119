## Connecting to the Database
Import the necessary packages and connect to the database **data.sqlite**.

```python
#Your code here
import sqlite3
import pandas as pd
connection = sqlite3.connect('data.sqlite')
cr = connection.cursor()
```

## Display the names of all the employees in Boston.

```python
#Your code here


cr.execute("""SELECT firstName FROM employees e
                       JOIN offices o
                       USING(officeCode)
                       WHERE city = 'Boston';""").fetchall()
```

## Do any offices have no employees?

```python
#Your code here
cr.execute("""SELECT city,
                    COUNT(*)
                    FROM offices
                    LEFT JOIN employees
                    using(officeCode)
                    group by 1;""").fetchall()
```
